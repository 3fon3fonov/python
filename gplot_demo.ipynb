# Using gnuplot via gplot.py in Jupyter

Importing and creation of a gnuplot instance for Jupyter.

```python
from gplot import Iplot
iplot = Iplot()
```

### Plot a gnuplot function
Plot commands are executed by calling the instance.

```python
iplot('sin(x)')
```

### Plot data array
Data array or lists can be also passed in the plot call. Note the extra comma after "pt 7".

```python
import numpy as np
x = np.arange(100.)
y = np.sin(0.1*x) + 0.1*np.random.randn(100)

iplot(x**2, y, 'pt 7,', [5000], [0.4], 'lt 3 pt 5 ps 2 t "a big point"')
```

### Plot a data file
Data files and unix pipes are possible as well and need extra quotes.

```python
iplot('"< seq 10" us 1:(log($1)):(0.5):1 w e pt 7 palette t ""')
```

### 2D plots
splot (and plot) are class methods.

```python
iplot.splot('[-4:4][-4:4] (sin(x)/x*sin(y)/y)**2')
```

### Available gnuplot styles

```python
iplot.test()
```

The output before were png files which can be download with a right click. There is also svg and html/js/canvas; this is experimental and needs gnuplot.js.

```python
iplot = Iplot(imgfile='html')
iplot.grid()("sin(x), cos(x), atan(x)*1.5,", [1,2,3,4,3,2,1], 'w lp')
```

```python
iplot = Iplot(imgfile='svg')
iplot.grid()("sin(x), cos(x), atan(x)*1.5,", [1,2,3,4,3,2,1], 'w lp')
```

For gnuplot 5+, hypertext is available in svg ang html.
